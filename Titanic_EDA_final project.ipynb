```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set Seaborn style for better visuals
sns.set_style("whitegrid")

# Load the Titanic dataset
titanic = sns.load_dataset("titanic")

# Display the first few rows
titanic.head()
```

```python
# Display dataset info
titanic.info()
```

```python
# Display summary statistics
titanic.describe(include="all")
```

```python
# Check for missing values
titanic.isnull().sum()
```

```python
# Handle missing values
titanic["age"].fillna(titanic["age"].median(), inplace=True)
titanic["embarked"].fillna(titanic["embarked"].mode()[0], inplace=True)
```

```python
# Plot survival distribution
plt.figure(figsize=(8, 6))
sns.countplot(x="survived", data=titanic)
plt.title("Distribution of Survival")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

```python
# Plot passenger class distribution
plt.figure(figsize=(8, 6))
sns.countplot(x="pclass", data=titanic)
plt.title("Distribution of Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.show()
```

```python
# Plot age distribution
plt.figure(figsize=(8, 6))
sns.histplot(titanic["age"], bins=30, kde=True)
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
# Plot fare distribution
plt.figure(figsize=(8, 6))
sns.histplot(titanic["fare"], bins=30, kde=True)
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Count")
plt.show()
```

```python
# Plot sex distribution
plt.figure(figsize=(8, 6))
sns.countplot(x="sex", data=titanic)
plt.title("Distribution of Sex")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()

```

```python
# Plot embarked distribution
plt.figure(figsize=(8, 6))
sns.countplot(x="embarked", data=titanic)
plt.title("Distribution of Port of Embarkation")
plt.xlabel("Port (C = Cherbourg, Q = Queenstown, S = Southampton)")
plt.ylabel("Count")
plt.show()
```

```python
# Plot survival by passenger class
plt.figure(figsize=(8, 6))
sns.barplot(x="pclass", y="survived", data=titanic)
plt.title("Survival Rate by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.show()
```

```python
# Plot survival by sex
plt.figure(figsize=(8, 6))
sns.barplot(x="sex", y="survived", data=titanic)
plt.title("Survival Rate by Sex")
plt.xlabel("Sex")
plt.ylabel("Survival Rate")
plt.show()
```

```python
# Plot survival by age
plt.figure(figsize=(8, 6))
sns.boxplot(x="survived", y="age", data=titanic)
plt.title("Age Distribution by Survival")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Age")
plt.show()
```

```python
# Plot survival by fare
plt.figure(figsize=(8, 6))
sns.boxplot(x="survived", y="fare", data=titanic)
plt.title("Fare Distribution by Survival")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Fare")
plt.ylim(0, 100)  # Limit y-axis to focus on common fares
plt.show()
```

```python
# Plot survival by embarked
plt.figure(figsize=(8, 6))
sns.barplot(x="embarked", y="survived", data=titanic)
plt.title("Survival Rate by Port of Embarkation")
plt.xlabel("Port (C = Cherbourg, Q = Queenstown, S = Southampton)")
plt.ylabel("Survival Rate")
plt.show()
```

```python
# Pair plot of numerical variables
sns.pairplot(titanic[["age", "fare", "sibsp", "parch", "survived"]], hue="survived")
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(titanic[["age", "fare", "sibsp", "parch", "pclass", "survived"]].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Plot survival by class and sex
plt.figure(figsize=(8, 6))
sns.catplot(x="pclass", y="survived", hue="sex", kind="bar", data=titanic)
plt.title("Survival Rate by Class and Sex")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.show()
```

```python
# Box plot for fare outliers
plt.figure(figsize=(8, 6))
sns.boxplot(y="fare", data=titanic)
plt.title("Box Plot of Fare")
plt.ylabel("Fare")
plt.show()
```
