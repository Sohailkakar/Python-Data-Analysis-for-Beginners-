```python
import seaborn as sns
print(sns.__version__)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sns.set_style("whitegrid")
```

```python
# Sample dataset
data = pd.DataFrame({
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'Sales': [200, 250, 300, 280, 320],
    'Expenses': [150, 180, 200, 190, 210],
    'Category': ['Electronics', 'Electronics', 'Clothing', 'Clothing', 'Books']
})
```

```python
# Count plot
plt.figure(figsize=(8, 6))
sns.countplot(x='Category', data=data)
plt.title("Count of Items by Category")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

```python
# Bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x='Category', y='Sales', data=data)
plt.title("Average Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales ($)")
plt.show()
```

```python
# Box plot
plt.figure(figsize=(8, 6))
sns.boxplot(x='Category', y='Expenses', data=data)
plt.title("Expenses Distribution by Category")
plt.xlabel("Category")
plt.ylabel("Expenses ($)")
plt.show()
```

```python
# Violin plot
plt.figure(figsize=(8, 6))
sns.violinplot(x='Category', y='Sales', data=data)
plt.title("Sales Distribution by Category")
plt.xlabel("Category")
plt.ylabel("Sales ($)")
plt.show()
```

```python
# Pair plot
sns.pairplot(data[['Sales', 'Expenses']], diag_kind='hist')
plt.show()
```

```python
# Correlation matrix
corr = data[['Sales', 'Expenses']].corr()

# Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Custom bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x='Category', y='Sales', data=data, palette='viridis')
plt.title("Sales by Category with Custom Palette")
plt.xlabel("Category")
plt.ylabel("Sales ($)")
plt.show()
```

```python
# Bar plot with hue
plt.figure(figsize=(8, 6))
sns.barplot(x='Month', y='Sales', hue='Category', data=data)
plt.title("Sales by Month and Category")
plt.xlabel("Month")
plt.ylabel("Sales ($)")
plt.legend(title='Category')
plt.show()
```

```python
# Facet grid
g = sns.FacetGrid(data, col='Category', height=4)
g.map(plt.scatter, 'Sales', 'Expenses')
g.add_legend()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set plot style
sns.set_style("whitegrid")

# Expanded and practical sales dataset
sales_data = pd.DataFrame({
    'Month': ['Jan', 'Jan', 'Feb', 'Feb', 'Mar', 'Mar'],
    'Category': ['Electronics', 'Clothing', 'Electronics', 'Clothing', 'Electronics', 'Clothing'],
    'Sales': [200, 150, 250, 180, 300, 200],
    'Expenses': [120, 100, 150, 130, 180, 140]
})

# Create boxplot
plt.figure(figsize=(9, 6))
sns.boxplot(
    x='Category',
    y='Sales',
    hue='Month',
    data=sales_data,
    order=['Electronics', 'Clothing'],     # consistent category order
    hue_order=['Jan', 'Feb', 'Mar'],       # chronological hue order
    palette='Set2'
)

# Add titles and labels
plt.title("Monthly Sales Distribution by Category", fontsize=14)
plt.xlabel("Product Category")
plt.ylabel("Sales in USD")
plt.legend(title='Month')
plt.tight_layout()
plt.show()
```
